1. Count Vowels

Write a function `count_vowels(s)` that takes a string and returns the number of vowels in it (`a`, `e`, `i`, `o`, `u` — both upper and lower case).

```python
>>> count_vowels("hello")
2
>>> count_vowels("HEllO World")
3

````python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
````

2 BMI Calculator

Write a function bmi(weight_kg, height_m) that returns the Body Mass Index (BMI) using the formula:

$
BMI = \frac{\text{weight (kg)}}{(\text{height (m)})^2}
$

Include doctests for values like:
	•	weight_kg=70, height_m=1.75 → 22.86 (rounded to two decimals)

````python
### BEGIN SOLUTION
def bmi(weight_kg, height_m):
    """
    Calculates the Body Mass Index (BMI).

    >>> round(bmi(70, 1.75), 2)
    22.86
    """
    return weight_kg / (height_m ** 2)
### END SOLUTION
````

````python
assert round(bmi(70, 1.75), 2) == 22.86
assert round(bmi(50, 1.6), 2) == 19.53
````

````python
def filter_short(words, max_len):
    """
    Returns a list of words whose length is less than or equal to max_len.

    >>> filter_short(['hi', 'hello', 'no', 'yes'], 2)
    ['hi', 'no']
    >>> filter_short(['apple', 'banana', 'pear'], 4)
    ['pear']
    >>> filter_short([], 5)
    []
    """
    return [word for word in words if len(word) <= max_len]
````

````python
new_words = ['apple', 'banana', 'kiwi', 'grapefruit', 'cat']
max_length = 5

result = filter_short(new_words, max_length)
print(f"Words shorter than or equal to {max_length} characters: {result}")
````

````python
def bmi(weight_kg, height_m):
    """
    Calculates the Body Mass Index (BMI) using the standard formula.

    The BMI is calculated as: weight (kg) / (height (m))^2

    Args:
        weight_kg (float or int): The weight of the person in kilograms.
        height_m (float or int): The height of the person in meters.

    Returns:
        float: The calculated Body Mass Index (BMI).

    Examples:
        >>> round(bmi(70, 1.75), 2)
        22.86
        >>> round(bmi(80, 1.80), 2)
        24.69
    """
    return weight_kg / (height_m ** 2)
````
