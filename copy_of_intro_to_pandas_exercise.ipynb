Q1. Upload and Preview Data

	1.	Upload the file airquality.csv to your Colab environment.\\

	2.	Load the first 10,000 rows into a Pandas dataframe called airquality.\\

	3.	Display the first 5 rows of the dataframe.\\

```python
import pandas as pd

# Define the path to the original Excel file
excel_file_path = 'us_air_quality_one_week_2022_06_23 (1).xlsx'
csv_file_path = 'airquality.csv'

# Load the Excel file
try:
    df_excel = pd.read_excel(excel_file_path)
    # Save the DataFrame to a proper CSV file
    df_excel.to_csv(csv_file_path, index=False, encoding='utf-8')
    print(f"Successfully converted '{excel_file_path}' to '{csv_file_path}'.")
except FileNotFoundError:
    print(f"Error: Excel file '{excel_file_path}' not found.")
    # If Excel file not found, assume airquality.csv is already uploaded and proceed with reading it
except Exception as e:
    print(f"An error occurred during Excel to CSV conversion: {e}")

# Load the first 10,000 rows of airquality.csv into a Pandas DataFrame
# Using 'latin1' encoding and 'python' engine for robustness, though a clean CSV should not need it.
airquality = pd.read_csv(csv_file_path, nrows=10000, encoding='latin1', engine='python')

# Display the first 5 rows of the DataFrame
print("\nFirst 5 rows of the airquality DataFrame:")
print(airquality.head())
```

```python
# 1. Use .info() and .describe(include="all") to explore the dataset.
print("\nDataFrame Information:")
airquality.info()

print("\nDataFrame Description (including all columns):")
display(airquality.describe(include='all'))

# 2. Identify any columns that have missing (NaN) values.
print("\nMissing values per column:")
missing_values = airquality.isnull().sum()
display(missing_values[missing_values > 0])

# 3. Drop all rows that contain missing values.
initial_rows = airquality.shape[0]
airquality.dropna(inplace=True)
final_rows = airquality.shape[0]
print(f"\nDropped {initial_rows - final_rows} rows with missing values.")
print(f"DataFrame now has {final_rows} rows.")
```

Q3. Data Visualization

	1. For the `pollutant` column, we have identified the unique values present in the data using `airquality['pollutant'].unique()` as ['bc' 'co']. Since there is no 'pm25' pollutant in the dataset, we will create visualizations for 'bc' and 'co'.

	2. Create a histogram of the 'bc' values.

	3. Create a boxplot for the 'bc' column.

	4. Create a histogram of the 'co' values.

	5. Create a boxplot for the 'co' column.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for 'bc' pollutant
bc_data = airquality[airquality['pollutant'] == 'bc']

# Create a histogram of 'bc' values
plt.figure(figsize=(10, 6))
sns.histplot(bc_data['value'], bins=30, kde=True)
plt.title('Histogram of Black Carbon (bc) Values')
plt.xlabel('Black Carbon Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create a boxplot for the 'bc' column
plt.figure(figsize=(10, 6))
sns.boxplot(y=bc_data['value'])
plt.title('Boxplot of Black Carbon (bc) Values')
plt.ylabel('Black Carbon Value')
plt.grid(True)
plt.show()
```

```python
# Filter data for 'co' pollutant
co_data = airquality[airquality['pollutant'] == 'co']

# Create a histogram of 'co' values
plt.figure(figsize=(10, 6))
sns.histplot(co_data['value'], bins=30, kde=True)
plt.title('Histogram of Carbon Monoxide (co) Values')
plt.xlabel('Carbon Monoxide Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create a boxplot for the 'co' column
plt.figure(figsize=(10, 6))
sns.boxplot(y=co_data['value'])
plt.title('Boxplot of Carbon Monoxide (co) Values')
plt.ylabel('Carbon Monoxide Value')
plt.grid(True)
plt.show()
```
