Q2. Data Summary and Cleaning

	1.	Use .info() and .describe(include="all") to explore the dataset.\\

	2.	Identify any columns that have missing (NaN) values.\\

	3.	Drop all rows that contain missing values.\\

```python
import pandas as pd

# Load the first 10,000 rows into a Pandas dataframe called airquality
airquality = pd.read_csv('airquality.csv', nrows=10000)

# Display the first 5 rows of the dataframe
airquality.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the DataFrame for 'co' pollutant
co_data = airquality[airquality['pollutant'] == 'co']

# Create a histogram of the 'co' values
plt.figure(figsize=(10, 6))
sns.histplot(co_data['value'], kde=True, bins=30)
plt.title('Histogram of CO Values')
plt.xlabel('CO Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the DataFrame for 'co' pollutant (if not already done)
co_data = airquality[airquality['pollutant'] == 'co']

# Create a boxplot for the 'co' column
plt.figure(figsize=(8, 6))
sns.boxplot(y=co_data['value'])
plt.title('Boxplot of CO Values')
plt.ylabel('CO Value')
plt.grid(axis='y', alpha=0.0)
plt.show()
```

```python
# Display unique values in the 'pollutant' column
print("Unique pollutants in the airquality DataFrame:")
print(airquality['pollutant'].unique())
```

```python
# Display unique values in the 'pollutant' column
print("Unique pollutants in the airquality DataFrame:")
print(airquality['pollutant'].unique())
```

```python
# Use .info() to explore the dataset
print('DataFrame Info:')
airquality.info()

# Use .describe(include="all") to explore the dataset
print('\nDataFrame Description (including all columns):')
airquality.describe(include="all")
```

```python
# Identify columns that have missing (NaN) values
print('Missing values before dropping:')
print(airquality.isnull().sum())

# Drop all rows that contain missing values
airquality.dropna(inplace=True)

print('\nMissing values after dropping:')
print(airquality.isnull().sum())
print('\nDataFrame Info after dropping NaN rows:')
airquality.info()
```

Q3. Data Visualization

	1.	Create a histogram of the pm25 values.\\

	2.	Create a boxplot for the pm25 column.\\


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the DataFrame for 'pm25' pollutant
pm25_data = airquality[airquality['pollutant'] == 'pm25']

# Create a histogram of the pm25 values
plt.figure(figsize=(10, 6))
sns.histplot(pm25_data['value'], kde=True, bins=30)
plt.title('Histogram of pm25 Values')
plt.xlabel('pm25 Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the DataFrame for 'pm25' pollutant (if not already done in the previous cell)
pm25_data = airquality[airquality['pollutant'] == 'pm25']

# Create a boxplot for the pm25 column
plt.figure(figsize=(8, 6))
sns.boxplot(y=pm25_data['value'])
plt.title('Boxplot of pm25 Values')
plt.ylabel('pm25 Value')
plt.grid(axis='y', alpha=0.0)
plt.show()
```
